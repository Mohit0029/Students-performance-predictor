```python

# 1. SETUP - IMPORT LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")

# ==============================================================================
# 2. LOAD DATA
# ==============================================================================
# Define the correct file path.
# This assumes your notebook is in the 'notebooks' folder.
file_path = '../data/student_performance_data.csv'

try:
    df = pd.read_csv(file_path)
    print("CSV file loaded successfully!")
except FileNotFoundError:
    print(f"Error: The file was not found at {file_path}")
    print("Please make sure the file name is correct and it's inside the 'data' folder.")

# ==============================================================================
# 3. INITIAL INSPECTION
# ==============================================================================
# Check if the dataframe 'df' was created before proceeding
if 'df' in locals():
    print("\n--- Data Overview ---")
    # Display the first 5 rows to get a feel for the data
    print("First 5 rows of the dataset:")
    display(df.head())

    # Get a summary of the dataframe, including data types and non-null counts
    print("\nDataset Info:")
    df.info()

    # Get a statistical summary of the numerical columns
    print("\nStatistical Summary:")
    display(df.describe())

    # Check for the total number of missing values in each column
    print("\nMissing Values Count:")
    print(df.isnull().sum())
    
    # Check for duplicate rows
    print(f"\nNumber of duplicate rows: {df.duplicated().sum()}")


# ==============================================================================
# 4. DATA CLEANING
# ==============================================================================
if 'df' in locals():
    # Make a copy to keep the original data intact
    df_cleaned = df.copy()

    # Drop duplicate rows if any exist
    df_cleaned.drop_duplicates(inplace=True)
    
    # For this example, we'll assume there are no missing values.
    # If there were, you would handle them here. For example:
    # median_score = df_cleaned['MathScore'].median()
    # df_cleaned['MathScore'].fillna(median_score, inplace=True)

    print("\n--- Data Cleaning Complete ---")
    print(f"Removed duplicates. Shape of data is now: {df_cleaned.shape}")


# ==============================================================================
# 5. DATA VISUALIZATION (EDA)
# ==============================================================================
if 'df' in locals():
    print("\n--- Starting Data Visualization ---")
    
    # Set the style for the plots
    sns.set_style("whitegrid")

    # Plot 1: Distribution of Final Scores
    plt.figure(figsize=(10, 6))
    sns.histplot(df_cleaned['FinalScore'], kde=True, bins=30)
    plt.title('Distribution of Final Scores')
    plt.xlabel('Final Score')
    plt.ylabel('Number of Students')
    plt.show()

    # Plot 2: Final Score vs. Study Hours per Day
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='StudyHoursPerDay', y='FinalScore', data=df_cleaned, alpha=0.6)
    plt.title('Final Score vs. Study Hours per Day')
    plt.xlabel('Study Hours per Day')
    plt.ylabel('Final Score')
    plt.show()
    
    # Plot 3: Final Scores by Parental Education
    plt.figure(figsize=(12, 7))
    sns.boxplot(x='ParentalEducation', y='FinalScore', data=df_cleaned, order=['High School', 'Graduate', 'Postgraduate', 'Doctorate'])
    plt.title('Final Scores by Parental Level of Education')
    plt.xlabel('Parental Education Level')
    plt.ylabel('Final Score')
    plt.xticks(rotation=45)
    plt.show()
```
